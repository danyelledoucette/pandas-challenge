```python
import pandas as pd
csv_path = "Resources/purchase_data.csv"

purchase_data = pd.read_csv(csv_path)
purchase_data.head()
```

```python
player_count = len(purchase_data["SN"].unique())

player_count_df = pd.DataFrame({"Total Players": [player_count]})
player_count_df
```

```python
unique_items = len(purchase_data["Item ID"].unique())

purchase_data['Purchase Price'] = purchase_data.iloc[:, 6:].sum(axis=1)
purchase_price = sum(purchase_data['Purchase Price'])/780

total_purchases = len(purchase_data)

total_revenue = sum(purchase_data['Purchase Price'])

purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items], 
                                       "Purchase Price": [purchase_price], 
                                       "Total Purchases": [total_purchases],
                                       "Total Revenue": [total_revenue]})
purchasing_analysis_df["Purchase Price"] = purchasing_analysis_df["Purchase Price"].astype(float).map(
    "${:,.2f}".format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].astype(float).map("${:,.2f}".format)

purchasing_analysis_df
```

```python
gender_demo = purchase_data.groupby(["Gender"])
gender_final = gender_demo["SN"].nunique()
gender_total = gender_final.sum()
gender_demo_df = pd.DataFrame({"Total Count": gender_final, "Percentage of Players": (gender_final/gender_total)*100})
gender_demo_df["Percentage of Players"] = gender_demo_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)
gender_demo_df
```

```python
gender_analysis = purchase_data.groupby(["Gender"])
gender_final = gender_analysis["SN"].nunique()
gender_total = gender_total.sum()

purchase_count = gender_analysis["Purchase ID"].count()
avg_purchase_price = gender_analysis["Price"].mean()
avg_purchase_total = gender_analysis["Price"].sum()
avg_purchase_per_person = avg_purchase_total/gender_final

gender_demographics = pd.DataFrame({"Purchase Count": purchase_count, 
                                    "Average Purchase Price": avg_purchase_price,
                                    "Total Purchase Value":avg_purchase_total,
                                    "Avg Purchase Total per Person": avg_purchase_per_person})

gender_demographics.index.name = "Gender"

gender_demographics.style.format({"Total Purchase Value":"${:,.2f}",
                                  "Average Purchase Price":"${:,.2f}",
                                  "Avg Purchase Total per Person":"${:,.2f}"})
```

```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins, labels=group_names)
purchase_data

age_group = purchase_data.groupby("Age Group")

age_count = age_group["SN"].nunique()

percentage_by_age = (age_count/player_count) * 100

age_demographics = pd.DataFrame({"Total Count": age_count, "Percentage of Players": percentage_by_age})

age_demographics.index.name = None

age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].astype(float).map("{:,.2f}%".format)

age_demographics
```

```python
purchase_by_age = age_group["Purchase ID"].count()

ave_purchase_by_group = age_group["Price"].mean()

total_purchase = age_group["Price"].sum()

ave_purchase_per_person = total_purchase/purchase_by_age

age_anaylsis = pd.DataFrame({"Purchase Count": purchase_by_age, 
                             "Average Purchase Price": ave_purchase_by_group, 
                             "Total Purchase Value": total_purchase, 
                             "Avg Total Purchase per Person": ave_purchase_per_person})

age_anaylsis["Total Purchase Value"] = age_anaylsis["Total Purchase Value"].astype(float).map("${:,.2f}".format)

age_anaylsis["Avg Total Purchase per Person"] = age_anaylsis["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

age_anaylsis["Average Purchase Price"] = age_anaylsis["Average Purchase Price"].astype(float).map("${:,.2f}".format)

age_anaylsis
```

```python
user_count=purchase_data.groupby(["SN"]).count()["Price"]

user_average=purchase_data.groupby(["SN"]).mean()["Price"].map("${0:,.2f}".format)

user_total=purchase_data.groupby(["SN"]).sum()["Price"].map("${0:,.2f}".format)

user_summary=pd.DataFrame({"Purchase Count":user_count,
                          "Average Purchase":user_average,
                          "Total Purchase Value":user_total})

user_summary.sort_values("Total Purchase Value",ascending=False).round(2).head()
```

```python
item_id = purchase_data.groupby(purchase_data['Item ID'])

unique_items = item_id['Item ID'].unique().str[0]

item_name = item_id['Item Name'].unique().str[0]

item_purchase_count = item_id['Age'].count()

item_price = item_id['Price'].unique().str[0]

item_purchase_total = item_id['Price'].sum()

item_summary =pd.DataFrame({'Item ID':unique_items,
                'Item Name':item_name,
                'Purchase Count':item_purchase_count,            
                'Item Price':item_price,
                'Total Purchase':item_purchase_total})


item_summary = item_summary.sort_values('Purchase Count', ascending=False)

item_summary_df = item_summary[['Item Name','Purchase Count', 'Item Price', 'Total Purchase']]

item_summary_df['Item Price'] = item_summary_df['Item Price'].astype(float).map("${:,.2f}".format)

item_summary_df['Total Purchase'] = item_summary_df['Total Purchase'].astype(float).map("${:,.2f}".format)

item_summary_df.head()
```

```python
most_profit=item_summary.sort_values('Total Purchase', ascending=False)

most_profit["Total Purchase"]=most_profit["Total Purchase"].astype(float).map("${:,.2f}".format)

most_profit["Item Price"]=most_profit["Item Price"].astype(float).map("${:,.2f}".format)

most_profit.drop(most_profit.columns[[0]], axis=1, inplace=True)

most_profit.head()
```
